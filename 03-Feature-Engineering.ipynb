```python
import pandas as pd
import numpy as np
import datetime as dt
pd.set_option('display.Max_rows', None)
```

```python
sales = pd.read_csv('cleaned-sales.csv')
```

```python
sales.sample(5)
```

```python
# info
sales.info()
```

```python
# check for duplicated
sales.duplicated().sum()
```

### Copy dataset

```python
df = sales.copy()
```

### Change dtypes 

```python
df['Order Date'].value_counts()
```

```python
# Order Date -> datetime
df['Order Date'] = pd.to_datetime(df['Order Date'], dayfirst = True)
```

```python
# Ship Date -> datetime
df['Ship Date'] = pd.to_datetime(df['Ship Date'], dayfirst = True)
```

```python
# Postal Code -> int
df['Postal Code'] = df['Postal Code'].astype('int')
```

```python
df.sample(2)
```

# Feature Engineering 

### 1. Order Date

```python
# Month order
df['Order Month'] = df['Order Date'].dt.month_name()
```

```python
# Order day
df['Order Day'] = df['Order Date'].dt.day_name()
```

### 2. Ship Month

```python
# Ship day
df['Ship day'] = df['Ship Date'].dt.day_name()
```

```python
df.head()
```

```python
# drop column
cols_to_drop = ['Order Date', 'Ship Date']
df.drop(columns = cols_to_drop, inplace = True)
```

```python
df.info()
```

### Export dataset

```python
df.to_csv('processed-salesdata.csv', index = False)
```
